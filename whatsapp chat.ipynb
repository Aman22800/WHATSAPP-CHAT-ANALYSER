```python
import re
import pandas as pd
```

```python
f=open('whatsapp2.txt','r',encoding='utf-8')
```

```python
data=f.read()
#print(data)
```

```python
pattern = r'\[\d{2}/\d{2}/\d{2},\s\d{2}:\d{2}:\d{2}\]'
```

```python
messages=re.split(pattern,data)[1:]
messages
```

```python
dates= re.findall(pattern,data)
dates
```

```python
df = pd.DataFrame({'user_message': messages, 'message_date': dates})

# Remove square brackets and convert to datetime
df['message_date'] = df['message_date'].str.strip("[]")  # Remove brackets
df['message_date'] = pd.to_datetime(df['message_date'], format="%d/%m/%y, %H:%M:%S")

# Rename column
df.rename(columns={'message_date': 'date'}, inplace=True)

# Display DataFrame
print(df.head())
```
